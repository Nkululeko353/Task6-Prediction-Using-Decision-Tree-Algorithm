```python
# load the training data 
iris_data = pd.read_csv(location)
iris_data
```

```python
print(iris_data)
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

X = iris_data[features]
y = iris_data['class']

print(X)
print(y)
```

```python
# There are many alternatives to visualize a Decision Tree

#1. We can fit a simple Decision Tree Classifier where we fit only the X and y variable without including the training and
# testing sets.


from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg



dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

import graphviz
from sklearn import tree
from sklearn import datasets
# import some data to play with
iris = datasets.load_iris()
dot_data = tree.export_graphviz(dtree, out_file=None,  feature_names=features,  
    class_names=iris.target_names,  filled=True, rounded=True, special_characters=True) 
graph = graphviz.Source(dot_data)  
graph.render('dtree_render',view=True)
```

```python
# A Decision Tree in now visualized.
```

```python
# A Simple Explaination of a visualized Decision Tree to aid a clear understanding.

# As illustrated on the Diagram

# petal_length â‰¤ 2.45 means that a petal length of 2.45 or lower will follow the True arrow (to the left), and the rest will follow the False arrow (to the right).

#gini = 0.667 refers to the quality of the split, and is always a number between 0.0 and 0.5, where 0.0 would mean all of the samples got the same result, and 0.5 would mean that the split is done exactly in the middle.

#samples = 150  means that there are 150 samples left at this point in the decision, which is all of them since this is the first step.

#value = [50, 50, 50]  means that of these 150 samples altogether , 50 will be classified correctly as "Setosa", and 50+50=100 will be classified as" versicolor".
```

```python
#Gini
#There are many ways to split the samples, we use the GINI method in this tutorial.

#The Gini method uses this formula:

#Gini = 1 - (x/n)2 - (y/n)2
#Where x is the number of positive answers, n is the number of samples, and y is the number of negative answers , which gives us this calculation:
```

```python
1 - (50 / 150)**2 -(50/150)**2-(50/150)**2
```

```python
# This is how we end up getting the gini of  0.667 as illustrated on the diagram
```

```python
# However we can choose the depth of the tree according to our own preference. The Decision Tree can be visualized by choosing 
# at at which depth the tree should stop.

HOW_DEEP_TREES = 2
clf = tree.DecisionTreeClassifier(random_state=0, max_depth=HOW_DEEP_TREES)
clf = clf.fit(X,y)
clf
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None,  feature_names=features,  
    class_names=iris.target_names,  filled=True, rounded=True, special_characters=True) 
graph = graphviz.Source(dot_data)  
graph.render('dtree_render',view=True)
```

```python
# However reducing the depth of the tree forms a crucial aspect by preveting overfiiting.
```

```python
#2. Another simple way to visualize a decision tree is by calling a simple tree plot method and this can be accomplished 
# by call the method as follows:
tree.plot_tree(clf);
# Here the tree is visualized ,it is not saved on a location.

# Another alternative by visualizing the decision tree by calling the simple tree plot method is as follows:
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

```python
# In the preceding examples we've visualized a decision tree where we include and fit the decision tree using only the
# x and y variable only without including the training and testing.

# 2. Another example can be when we are fitting the decision tree with a training set
# First we will split the data into training and test data
# So we first import using the following 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

```python
print(X_train.shape)
```

```python
y.unique()
```

```python
# Take note of the following:
# Here we use the Entropy method to split a tree
# We've fitted a decision tree using X_train and Y_train
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X_train, y_train)
```

```python
clf
```

```python
import graphviz
from sklearn import tree
from sklearn import datasets
# import some data to play with
iris = datasets.load_iris()
dot_data = tree.export_graphviz(clf, out_file=None,  feature_names=features,  
    class_names=iris.target_names,  filled=True, rounded=True, special_characters=True) 
graph = graphviz.Source(dot_data)  
graph.render('dtree_render',view=True)
```

```python


# Again here 
# A Simple Explaination of a visualized Decision Tree to aid a clear understanding.

# # As illustrated on the Diagram

# petal_length â‰¤ 2.45 means that a petal length of 2.45 or lower will follow the True arrow (to the left), 
#and the rest will follow the False arrow (to the right).
#  entropy =  1.581
#  refers to the quality of the split, and is always a number between 0.0 and 0.5, 
# where 0.0 would mean all of the samples got the same result, and 0.5 would mean that the split is done exactly in the middle.
#  #samples = 105  means that there are 105 samples left at this point in the decision, which is all of them since this is the first step.
# value = [35, 38, 32] means that of these 150 samples altogether , 35 will be classified correctly as "Setosa", and 38+32=70 will be classified as" versicolor".
```

```python
# Here we've splitted the tree using the entropy method,and after running the code we get the entropy of 
#1.5814250309620266 as shown on the visualised decision tree.
-1*( ((35.0/105)*np.log2(35.0/105))+(38.0/105)*np.log2(38.0/105)+(32/105)*np.log2(32/105))
```

```python
# After being fitted, the model can then be used to predict the class of samples:

y_pred = clf.predict(X_test)
```

```python
# import libraries for metrics and reporting
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn import metrics
from sklearn.metrics import classification_report
# print the confusion metrics
metrics.confusion_matrix(y_test, y_pred)
```

```python
# accuracy score
metrics.accuracy_score(y_test, y_pred)
```

```python
# classification report
print(classification_report(y_test, y_pred))
```
